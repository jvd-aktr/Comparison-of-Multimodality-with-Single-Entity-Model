```python
import os
import requests
import openpyxl
import time

def saveNotDownloadedIndex(listofIndex):
    notDownloadedBook = openpyxl.load_workbook('notDownloadedNeg.xlsx')
    notDownloadedSheet = notDownloadedBook['Sheet1']
    column = 1

    max_row = notDownloadedSheet.max_row

    for i, value in enumerate(listofIndex):
        notDownloadedSheet.cell(row=max_row+i+1, column=column).value = value
    notDownloadedBook.save('notDownloadedNeg.xlsx')

workbook = openpyxl.load_workbook('/Users/abdurrazzak/Razzak/Spring23/AML/Project/archive/codes/negImg.xlsx')
worksheet = workbook['Sheet1']

file_path = '/Users/abdurrazzak/Razzak/Spring23/AML/Project/archive/codes/images/neg/'
label = '0'
rowNum = 0;
unableToDownload = []
print('Going to download the neg class images')
for row in worksheet.iter_rows():
    if rowNum !=0:
        url = row[0].value
        try:
            response = requests.get(url)
            if(response.status_code == 200):
                ext = label + '_' + str(rowNum) + '.jpg'
                joinedPath = os.path.join(file_path, ext)
#                 print(joinedPath)
                with open(joinedPath, 'wb') as f:
                    f.write(response.content)
                print('file saved: ', ext)
            else:
                unableToDownload.append(rowNum)
                print('Couldnot Get the file, notDownloaded size: ', len(unableToDownload))
                if len(unableToDownload) >= 10 :
                    saveNotDownloadedIndex(unableToDownload)
                    print('Saved row numbers of UnableToDownload: ', len(unableToDownload))
                    unableToDownload = []
                    time.sleep(2)
                    
        except:
            unableToDownload.append(rowNum)
            print('Couldnot established connection, notDownloaded size: ', len(unableToDownload))
            if len(unableToDownload) >= 10:
                saveNotDownloadedIndex(unableToDownload)
                print('Saved row numbers of UnableToDownload: ', len(unableToDownload))
                unableToDownload = []
                time.sleep(2)
    time.sleep(5)
    rowNum+=1
print('---------DONE----------')
print(unableToDownload)
```
